```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
```

#### Get Data

```python
df = pd.read_csv('grouped_cases.csv')
df
```

#### How do total cases trend over time?

```python
fig, ax = plt.subplots()
ax.plot(df['DATE'], df['TOTAL CASES'], color='mediumseagreen')

# Set axis ticks
ax.set_xticks(ax.get_xticks()[::2])
ax.set_xticklabels(df['DATE'][::2], rotation = 45)
ax.get_yaxis().set_major_formatter(ticker.FuncFormatter(lambda x, p: format(int(x), ',')))

# Set title
ax.set_title('Total COVID-19 Cases')
ax.title.set_weight('bold')
ax.title.set_size(14)
ax.title.set_position([.5, 1.08])

# Set spines
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.spines['bottom'].set_lw(1.5)
ax.spines['left'].set_lw(1.5)

fig.show()
```

#### Compare total cases and cumulative incidence

```python
fig, ax1 = plt.subplots(figsize=(16,8))
ax2 = ax1.twinx()

ax1.plot(df['DATE'], df['TOTAL CASES'], color='goldenrod', label='Total Cases')
ax2.plot(df['DATE'], df['BLACK CI'], color='firebrick', label='Cumulative Incidence (Black)')
ax2.plot(df['DATE'], df['HISPANIC CI'], color='mediumseagreen', label='Cumulative Incidence (Hispanic)')
ax2.plot(df['DATE'], df['WHITE CI'], color='royalblue', label='Cumulative Incidence (White)')

# set title and axis labels
ax1.set_title('Total Cases vs. Cumulative Incidence by Race')
ax1.title.set_weight('bold')
ax1.title.set_size(14)
ax1.title.set_position([.5, 1.08])
ax1.set_ylabel('Total Cases', labelpad=10)
ax2.set_ylabel('Cumulative Incidence', labelpad=15)

# set xticks
ax1.set_xticks(ax1.get_xticks()[::2])
ax1.set_xticklabels(df['DATE'][::2], rotation = 45)
    
# set spines
ax1.spines['right'].set_lw(1.5)
ax1.spines['top'].set_visible(False)
ax1.spines['bottom'].set_lw(1.5)
ax1.spines['left'].set_lw(1.5)
ax2.spines['right'].set_lw(1.5)
ax2.spines['top'].set_visible(False)
ax2.spines['bottom'].set_lw(1.5)
ax2.spines['left'].set_lw(1.5)

# set legend
fig.legend(loc='upper left')

fig.show()
```

#### How do trends in cumulative incidence vary by race?

```python
fig, ax = plt.subplots(3, 1)
ax[0].plot(df['DATE'], df['BLACK CI'], color='mediumseagreen')
ax[1].plot(df['DATE'], df['HISPANIC CI'], color='firebrick')
ax[2].plot(df['DATE'], df['WHITE CI'], color='royalblue')
```

```python
y_list = ['BLACK CI', 'HISPANIC CI', 'WHITE CI']
colors = ['mediumseagreen', 'firebrick', 'royalblue']
labels = ['Black Cumulative Incidence', 'Hispanic Cumulative Incidence', 'White Cumulative Incidence']

fig, ax = plt.subplots(3, 1, sharex='all', sharey='all')
for i in range(len(y_list)):
    ax[i].plot(df['DATE'], df[y_list[i]], color=colors[i], label=labels[i])
    
# set title and axis labels
ax[0].set_title('Cumulative Incidence by Race')
ax[0].title.set_weight('bold')
ax[0].title.set_size(14)
ax[0].title.set_position([.5, 1.08])
ax[1].set_ylabel('Cumulative Incidence', labelpad=15)

# set xticks
ax[2].set_xticks(ax[0].get_xticks()[::2])
ax[2].set_xticklabels(df['DATE'][::2], rotation = 45)
    
fig.legend(loc='center right')
    
fig.show()
```
